### Women's hoodies and sweatshirts

```python
prod_name_new = []
prod_rating_new = []
prod_price_new = []
prod_id_new = []

d = {}

i = 1

while i < 3:
    url = "https://www.amazon.com/Best-Sellers-Clothing-Shoes-Jewelry-Womens-Fashion-Hoodies-Sweatshirts/zgbs/fashion/1258603011/ref=zg_bs_pg_1_fashion?_encoding=UTF8&pg={}".format(i)
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url)

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    wait = WebDriverWait(driver, 10)
    
    time.sleep(10)
    
    html = driver.page_source
    soup = BeautifulSoup(html)
    
    
    prod_name = soup.find_all('div',attrs={'class':'_cDEzb_p13n-sc-css-line-clamp-3_g3dy1'})
    prod_price = soup.find_all('span',attrs={'class':'_cDEzb_p13n-sc-price_3mJ9Z'})
    prod_price.extend(soup.find_all('span',attrs={'class':'p13n-sc-price'}))
    prod_rating = soup.find_all('span',attrs={'class':'a-icon-alt'})
    prod_div = soup.find_all('div',attrs={'class':'p13n-sc-uncoverable-faceout'})
    tmp_prod_id = [x.find('a')['href'] for x in prod_div]
    prod_id = [re.search(r'\bB0[a-zA-Z0-9]*\b', x).group() for x in tmp_prod_id]


    temp_name = []
    temp_rating = []
    temp_price = []
    temp_id = []

    for r in range(len(prod_name)):
        temp_name.append(prod_name[r].text)

    for r in range(len(prod_rating)):
        temp_rating.append(prod_rating[r].text)

    for r in range(len(prod_price)):
        temp_price.append(prod_price[r].text)
        
    temp_id = prod_id


    prod_name_new.extend(temp_name)
    prod_rating_new.extend(temp_rating)
    prod_price_new.extend(temp_price)
    prod_id_new.extend(temp_id)
    
    i = i+1
    
    driver.close()
    
prod_price_new = prod_price_new[:-1]
prod_id_new = prod_id_new[:-1]

prod_rating_new = [float(x.split(' ')[0]) for x in prod_rating_new]
prod_price_new = [float(x.replace('$','')) for x in prod_price_new]

d['Product Name'] = prod_name_new
d['Product Rating'] = prod_rating_new
d['Product Price'] = prod_price_new
d['Product ID'] = prod_id_new

hoodies = pd.DataFrame(d)
```

## Comibining and creating ADS for modelling

```python

hoodies = pd.read_csv('hoodies.csv')
jeans = pd.read_csv('jeands.csv')

hoodie_colors = pd.read_csv('hoodies_colors.csv')
jeans_colors = pd.read_csv('jeans_colors.csv')
```

```python
hoodies
```

```python
all_apparel = pd.concat([hoodies,jeans], ignore_index=True)
all_apparel.shape, all_apparel['Product ID'].nunique()
```

```python
all_colors = pd.concat([hoodie_colors,jeans_colors], ignore_index=True)
all_colors.shape, all_colors['Product ID'].nunique()
```

```python
merged = all_apparel.merge(all_colors, on='Product ID')
```

```python
merged
```

```python
# Create features related to product categories and type

merged['Hoodies'] = np.where((merged['Product Name'].str.lower().str.contains('hoodie')) | 
                             (merged['Product Name'].str.lower().str.contains('sweatshirt')),1,0)

merged['Fleece'] = np.where(merged['Product Name'].str.lower().str.contains('fleece'),1,0)

merged['Pockets'] = np.where(merged['Product Name'].str.lower().str.contains('pocket'),1,0)

merged['Zipper'] = np.where(merged['Product Name'].str.lower().str.contains('zip'),1,0)

merged['Cropped'] = np.where(merged['Product Name'].str.lower().str.contains('crop'),1,0)

merged['Oversize'] = np.where(merged['Product Name'].str.lower().str.contains('oversize'),1,0)

merged['Off shoulder'] = np.where(merged['Product Name'].str.lower().str.contains('off the shoulder'),1,0)

merged['Crewneck'] = np.where(merged['Product Name'].str.lower().str.contains('crewneck'),1,0)

merged['Jeans'] = np.where(merged['Product Name'].str.lower().str.contains('jean'),1,0)

merged['Leggings'] = np.where(merged['Product Name'].str.lower().str.contains('legging'),1,0)

merged['Jeggings'] = np.where(merged['Product Name'].str.lower().str.contains('jegging'),1,0)

merged['Bootcut'] = np.where(merged['Product Name'].str.lower().str.contains('bootcut'),1,0)

merged['Skinny'] = np.where(merged['Product Name'].str.lower().str.contains('skinny'),1,0)

merged['High Rise'] = np.where((merged['Product Name'].str.lower().str.contains('high rise')) | 
                             (merged['Product Name'].str.lower().str.contains('high waist')) | 
                             (merged['Product Name'].str.lower().str.contains('high-rise')),1,0)

merged['Baggy'] = np.where(merged['Product Name'].str.lower().str.contains('baggy'),1,0)

merged['Capri'] = np.where(merged['Product Name'].str.lower().str.contains('capri'),1,0)
```

## Regression

<p> To understand what the Product rating depends on and if we can identify any patterns in terms of colors/material</p>

```python
merged.columns
```

```python
feats = merged.drop(['Product Name','Product Rating','Product ID'],axis=1).columns.tolist()
```

```python
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_predict

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
from sklearn.metrics import r2_score, mean_absolute_error as mae


X = merged[feats]
y = merged['Product Rating']

X_train, X_holdout, y_train, y_holdout = train_test_split(X, y, test_size=0.2, random_state = 17)

X_train_copy = X_train.copy()

scaling_features = StandardScaler()
scaling_fit = scaling_features.fit(X_train[feats])
X_train_scaled = scaling_fit.transform(X_train[feats])
X_train_scaled = pd.DataFrame(X_train_scaled)
X_train_scaled.columns = X_train[feats].columns
X_train = X_train_scaled

X_holdout_scaled = scaling_fit.transform(X_holdout[feats])
X_holdout_scaled = pd.DataFrame(X_holdout_scaled)
X_holdout_scaled.columns = X_holdout[feats].columns
X_holdout = X_holdout_scaled

model = Lasso(alpha=0.01, random_state = 17).fit(X_train,y_train)

print(r2_score(y_train, model.predict(X_train)), r2_score(y_holdout, model.predict(X_holdout)))
imp = pd.DataFrame(data={'features':X_train.columns,
                   'coef': model.coef_
                  }).sort_values(by='coef',ascending=False)
print(len(np.round(imp[imp['coef']!=0],5)))
```

<h2> Word Count separately for jeans and hoodies</h2>

<p>Word Count on Product name to understand trending items in the Bestseller category of Jeans and Hoodies</p>

```python
from collections import Counter
product_desc_series = hoodies['Product Name']

# Combine all product descriptions into a single string
all_product_desc = ' '.join(product_desc_series)

# Tokenize the string into words
words = all_product_desc.split()

# Use Counter to count the occurrences of each word
word_counts = Counter(words)
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Print the word frequencies in descending order
for word, count in sorted_word_counts:
    print(f"{word}: {count} times")
pd.DataFrame(sorted_word_counts,columns=['Keyword','Occurence']).to_csv('Amazon_bestseller_hoodies_keywords.csv')
```
